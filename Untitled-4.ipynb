```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display

# Paths
data_path = "D:/processed_audio_clean/"
output_path = "D:/outputs/"

# Create directories if not exist
os.makedirs(output_path, exist_ok=True)

# Function to load audio and extract features
def extract_features(file_path, sr=16000, n_mfcc=13):
    """
    Load audio file and extract MFCC features.
    :param file_path: Path to the audio file
    :param sr: Sampling rate
    :param n_mfcc: Number of MFCC features to extract
    :return: MFCC features
    """
    # Load audio
    audio, _ = librosa.load(file_path, sr=sr, mono=True)
    # Extract MFCC features (using keyword arguments)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    return mfccs

# Example: Process all files in the data directory
audio_files = [f for f in os.listdir(data_path) if f.endswith('.wav')]

for audio_file in audio_files:
    try:
        print(f"Processing: {audio_file}")
        file_path = os.path.join(data_path, audio_file)
        mfccs = extract_features(file_path)
        
        # Visualize MFCC features
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(mfccs, x_axis='time', sr=16000)
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'MFCCs: {audio_file}')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error processing {audio_file}: {e}")
```

```python
def segment_audio(audio, sr, segment_duration=1.0, overlap=0.5):
    """
    Segment the audio into fixed durations with overlap.
    :param audio: Audio data (1D numpy array)
    :param sr: Sampling rate
    :param segment_duration: Duration of each segment in seconds
    :param overlap: Overlap fraction between consecutive segments
    :return: List of audio segments
    """
    segment_samples = int(segment_duration * sr)  # Convert duration to samples
    step = int(segment_samples * (1 - overlap))  # Step size between segments
    segments = []
    
    for start in range(0, len(audio) - segment_samples + 1, step):
        end = start + segment_samples
        segments.append(audio[start:end])
    
    return segments

# Function to process audio files
def process_and_segment_audio(file_path, sr=16000, segment_duration=1.0, overlap=0.5):
    """
    Load audio and split it into segments.
    :param file_path: Path to the audio file
    :param sr: Sampling rate
    :param segment_duration: Duration of each segment in seconds
    :param overlap: Overlap fraction between consecutive segments
    :return: List of audio segments
    """
    # Load audio
    audio, _ = librosa.load(file_path, sr=sr, mono=True)
    # Segment audio
    segments = segment_audio(audio, sr, segment_duration, overlap)
    return segments
```

```python
def extract_features_from_segments(segments, sr=16000, n_mfcc=13):
    """
    Extract MFCC features for a list of audio segments.
    :param segments: List of audio segments (numpy arrays)
    :param sr: Sampling rate
    :param n_mfcc: Number of MFCC features to extract
    :return: List of feature arrays (2D: time steps × n_mfcc)
    """
    features = []
    for i, segment in enumerate(segments):
        # Extract MFCCs for the segment
        mfcc = librosa.feature.mfcc(y=segment, sr=sr, n_mfcc=n_mfcc)
        features.append(mfcc.T)  # Transpose to time steps × n_mfcc
    return features

# Test with one audio file
file_path = os.path.join(data_path, audio_files[0])  # First audio file
segments = process_and_segment_audio(file_path)  # Segment the audio
features = extract_features_from_segments(segments)

print(f"Number of segments: {len(features)}")
print(f"Shape of one feature array: {features[0].shape}")  # Should be time steps × n_mfcc
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM, Dense, BatchNormalization

# Define the LSTM model
def build_speaker_embedding_model(input_shape, embedding_dim=128):
    """
    Build an LSTM-based neural network for speaker embeddings.
    :param input_shape: Shape of input features (time steps × n_mfcc)
    :param embedding_dim: Dimension of the output embedding
    :return: Keras model
    """
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=input_shape, activation='tanh'),
        LSTM(64, return_sequences=False, activation='tanh'),
        Dense(embedding_dim, activation='linear'),  # Output embeddings
        BatchNormalization()  # Normalize embeddings
    ])
    return model

# Example: Create the model
input_shape = (features[0].shape[0], features[0].shape[1])  # Time steps × n_mfcc
embedding_dim = 128
model = build_speaker_embedding_model(input_shape, embedding_dim)

# Display model architecture
model.summary()
```

```python
# Convert features to a 3D array (segments × time steps × n_mfcc)
X = np.array([f for f in features])  # Shape: (num_segments, time_steps, n_mfcc)

# Generate embeddings
embeddings = model.predict(X)

print(f"Generated embeddings: {embeddings.shape}")  # Shape: (num_segments, embedding_dim)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np

# Normalize embeddings
scaler = StandardScaler()
normalized_embeddings = scaler.fit_transform(embeddings)

# Determine the number of clusters (if not known)
def find_optimal_clusters(embeddings, max_clusters=10):
    """
    Use the Elbow Method to find the optimal number of clusters.
    :param embeddings: Speaker embeddings
    :param max_clusters: Maximum number of clusters to test
    :return: Optimal number of clusters
    """
    distortions = []
    for k in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=0)
        kmeans.fit(embeddings)
        distortions.append(kmeans.inertia_)
    
    # Plot distortions
    plt.figure(figsize=(8, 4))
    plt.plot(range(2, max_clusters + 1), distortions, marker='o')
    plt.title("Elbow Method for Optimal Clusters")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Distortion")
    plt.show()

# Find optimal clusters
find_optimal_clusters(normalized_embeddings)

# Apply K-means with a chosen number of clusters (e.g., 3 speakers)
n_clusters = 3  # Replace with your estimated number of speakers
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(normalized_embeddings)

print(f"Cluster labels: {labels}")
```

```python
from sklearn.decomposition import PCA

# Reduce dimensionality to 2D for visualization
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(normalized_embeddings)

# Plot clusters
plt.figure(figsize=(8, 6))
for i in range(n_clusters):
    cluster_points = reduced_embeddings[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Speaker {i + 1}")
plt.title("Speaker Clustering")
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.legend()
plt.show()
```

```python
def generate_diarization_timeline(labels, segment_duration=1.0, overlap=0.5, sr=16000):
    """
    Generate a timeline showing speaker labels for each segment.
    :param labels: Cluster labels (one per segment)
    :param segment_duration: Duration of each segment in seconds
    :param overlap: Overlap fraction between segments
    :return: List of (start_time, end_time, speaker_label) tuples
    """
    step = segment_duration * (1 - overlap)  # Step size between segments
    timeline = []
    for i, label in enumerate(labels):
        start_time = i * step
        end_time = start_time + segment_duration
        timeline.append((start_time, end_time, f"Speaker {label + 1}"))
    return timeline

# Generate the timeline
timeline = generate_diarization_timeline(labels)

# Display the timeline
for start, end, speaker in timeline:
    print(f"{start:.2f}s - {end:.2f}s: {speaker}")
```

```python
import matplotlib.patches as mpatches

def plot_diarization_timeline(timeline, audio_duration):
    """
    Plot the diarization timeline.
    :param timeline: List of (start_time, end_time, speaker_label) tuples
    :param audio_duration: Total duration of the audio in seconds
    """
    fig, ax = plt.subplots(figsize=(10, 2))
    colors = plt.cm.tab20.colors  # Use a colormap for speakers
    speaker_colors = {}
    
    for start, end, speaker in timeline:
        if speaker not in speaker_colors:
            speaker_colors[speaker] = colors[len(speaker_colors) % len(colors)]
        ax.add_patch(mpatches.Rectangle((start, 0), end - start, 1,
                                         color=speaker_colors[speaker],
                                         label=speaker if speaker not in ax.get_legend_handles_labels()[1] else ""))
    
    ax.set_xlim(0, audio_duration)
    ax.set_ylim(0, 1)
    ax.set_xlabel("Time (s)")
    ax.set_yticks([])
    ax.set_title("Speaker Diarization Timeline")
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3)
    plt.tight_layout()
    plt.show()

# Example: Visualize the diarization timeline
# Define segment duration and overlap
segment_duration = 1.0  # Segment length in seconds
overlap = 0.5  # Overlap fraction between segments

# Calculate total audio duration
audio_duration = len(segments) * (segment_duration * (1 - overlap))

# Visualize the diarization timeline
plot_diarization_timeline(timeline, audio_duration)
```

```python
import csv

def save_timeline_to_csv(timeline, output_path="D:/outputs/diarization_results.csv"):
    """
    Save diarization timeline to a CSV file.
    :param timeline: List of (start_time, end_time, speaker_label) tuples
    :param output_path: Path to save the CSV file
    """
    with open(output_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Start Time (s)", "End Time (s)", "Speaker Label"])
        writer.writerows(timeline)
    print(f"Diarization results saved to {output_path}")

# Save the timeline to a CSV file
save_timeline_to_csv(timeline)
```

```python
def compute_der(predicted, ground_truth):
    """
    Compute Diarization Error Rate (DER).
    :param predicted: List of predicted segments (start_time, end_time, speaker)
    :param ground_truth: List of ground truth segments (start_time, end_time, speaker)
    :return: DER, False Alarm, Missed Speech, Speaker Confusion
    """
    total_time = 0
    false_alarm = 0
    missed_speech = 0
    speaker_confusion = 0

    for gt_segment in ground_truth:
        gt_start, gt_end, gt_speaker = gt_segment["start_time"], gt_segment["end_time"], gt_segment["speaker"]
        total_time += (gt_end - gt_start)

        # Find overlapping predicted segments
        overlaps = [
            pred for pred in predicted
            if not (pred[1] <= gt_start or pred[0] >= gt_end)  # Check time overlap
        ]

        # If no overlaps, it's a missed speech
        if not overlaps:
            missed_speech += (gt_end - gt_start)
        else:
            for pred_start, pred_end, pred_speaker in overlaps:
                overlap_start = max(gt_start, pred_start)
                overlap_end = min(gt_end, pred_end)
                overlap_duration = max(0, overlap_end - overlap_start)

                if pred_speaker != gt_speaker:
                    speaker_confusion += overlap_duration

        # False alarms: Predicted segments outside ground truth
        for pred_start, pred_end, pred_speaker in overlaps:
            if pred_start < gt_start or pred_end > gt_end:
                false_alarm += max(0, min(pred_end, gt_end) - max(pred_start, gt_start))

    der = (false_alarm + missed_speech + speaker_confusion) / total_time
    return der, false_alarm, missed_speech, speaker_confusion

# Example: Compute DER
der, false_alarm, missed_speech, speaker_confusion = compute_der(timeline, ground_truth)
print(f"DER: {der:.2%}")
print(f"False Alarm: {false_alarm:.2f}s, Missed Speech: {missed_speech:.2f}s, Speaker Confusion: {speaker_confusion:.2f}s")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def create_confusion_matrix(predicted, ground_truth):
    """
    Generate a confusion matrix comparing predicted vs ground truth speakers.
    :param predicted: List of predicted speaker labels
    :param ground_truth: List of ground truth speaker labels
    """
    pred_labels = []
    gt_labels = []

    for pred, gt in zip(predicted, ground_truth):
        pred_labels.append(pred[2])  # Extract speaker from predicted
        gt_labels.append(gt["speaker"])  # Extract speaker from ground truth

    # Generate confusion matrix
    cm = confusion_matrix(gt_labels, pred_labels, labels=list(set(gt_labels + pred_labels)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(set(gt_labels + pred_labels)))
    disp.plot(cmap="Blues", xticks_rotation=45)
    plt.title("Speaker Confusion Matrix")
    plt.show()

# Example: Create confusion matrix
create_confusion_matrix(timeline, ground_truth)
```

```python
import numpy as np
import librosa
import librosa.display
import soundfile as sf
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load mono audio file
audio_file = "D:/processed_audio_clean/Cam10-Test1-Section1.wav"
y, sr = librosa.load(audio_file, sr=16000, mono=True)

# Perform Short-Time Fourier Transform (STFT)
stft = librosa.stft(y, n_fft=1024, hop_length=512)
magnitude, phase = np.abs(stft), np.angle(stft)

# Reshape magnitude into 2D array for clustering
time_frames, frequency_bins = magnitude.shape
magnitude_reshaped = magnitude.T.flatten().reshape(-1, 1)  # Shape: (time_frames × frequency_bins, 1)

# Apply K-means clustering to separate 3 sources (e.g., speakers)
n_clusters = 3  # Number of speakers
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(magnitude_reshaped)

# Reshape labels back to original spectrogram shape
labels = kmeans.labels_.reshape(frequency_bins, time_frames).T  # Shape: (time_frames, frequency_bins)

# Create masks for each speaker based on clustering
masks = np.zeros((n_clusters, time_frames, frequency_bins), dtype=np.float32)
for i in range(n_clusters):
    masks[i] = (labels == i).astype(np.float32)

# Separate each speaker using masks
separated_signals = []
for i in range(n_clusters):
    masked_magnitude = magnitude * masks[i]
    reconstructed_stft = masked_magnitude * np.exp(1j * phase)  # Combine with original phase
    separated_signal = librosa.istft(reconstructed_stft, hop_length=512)
    separated_signals.append(separated_signal)
    output_file = f"D:/speaker_{i + 1}_aci.wav"
    sf.write(output_file, separated_signal, sr)
    print(f"Speaker {i + 1} signal saved to {output_file}")

# Visualize the STFT magnitude spectrogram of the original and separated signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
librosa.display.specshow(librosa.amplitude_to_db(magnitude, ref=np.max), sr=sr, hop_length=512, y_axis='log', x_axis='time')
plt.title("Original Spectrogram")
for i, separated_signal in enumerate(separated_signals):
    plt.subplot(2, 2, i + 2)
    spec = np.abs(librosa.stft(separated_signal, n_fft=1024, hop_length=512))
    librosa.display.specshow(librosa.amplitude_to_db(spec, ref=np.max), sr=sr, hop_length=512, y_axis='log', x_axis='time')
    plt.title(f"Speaker {i + 1} Spectrogram")
plt.tight_layout()
plt.show()
```
